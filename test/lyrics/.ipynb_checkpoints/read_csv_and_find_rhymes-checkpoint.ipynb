```python
def split_vowel(str1):
    vowel = ['æ', 'ə', 'ɑ', 'ɔ', 'o', 'a', 'e', 'ɛ', 'i', 'ɪ', 'ʊ', 'u', 'ʌ']
    ipa = p.convert(str1, keep_punct=False, stress_marks=None)
    result = []
    iters = 0
    tolerate_flag = False 
    while True:
        iters += 1
        if iters > len(ipa):
            break
            
        if ipa[-iters] not in vowel:
            if not tolerate_flag:
                result.append(ipa[-iters])
            else:
                break
        else:
            result.append(ipa[-iters])
            if not tolerate_flag:
                tolerate_flag = True
                continue
            
    return ''.join(result[: : -1])
```

```python
f = csv.reader(open('Slaughter.csv', encoding='utf-8'))
rhymes = []
bands = []
lyrics = []
lyrics_dict = dict()
for lyric in f:
    rhyme = split_vowel(lyric[0].split()[-1])
    rhymes.append(rhyme)
    lyrics.append(lyric[0])
bands = ['Slaughter' for i in range(len(lyrics))]
lyrics_dict.update({'bands': bands, 'lyrics': lyrics, 'rhymes': rhymes})
```

```python
df = pd.DataFrame(lyrics_dict)
df
```

```python
artists = ['Guns N\' Roses', 'Black Sabbath']
lyrics_dict = dict()
for artist in artists:
    f = csv.reader(open(f'{artist}.csv', encoding='utf-8'))

    for lyric in f:
        rhyme = split_vowel(lyric[0].split()[-1])
        lyrics_dict.update({(artist, lyric[0]): rhyme})
lyrics_dict
```

```python
lyrics_dict_sort = sorted(lyrics_dict.items(), key=lambda x : x[1], reverse=False)
lyrics_dict_sort
```

```python
header = []
for head in sorted(lyrics_dict.keys()):
    header.append(head)
with open('./lyrics.csv', 'w', encoding='utf-8',newline='') as f:
    writer = csv.DictWriter(f, fieldnames=header)
    writer.writerow(lyrics_dict)
```
